### EXTRAE LINK

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import time

# Inicializar el navegador (asegúrate de tener el driver correspondiente instalado y en el PATH)
driver = webdriver.Chrome()

# La URL de la página web
url = "https://www.genesys-pgr.org/a/images/v2ZW8lQwlep"

# Abre la página en el navegador
driver.get(url)

# Espera a que aparezca el elemento esperado antes de continuar
elemento_esperado = WebDriverWait(driver, 80).until(
    EC.presence_of_element_located((By.XPATH, '//*[@id="the-app"]/div/div/div[2]/div/div[2]/div/div[4]/div/div[2]/div/div'))
)

# Contenedor para almacenar enlaces
enlaces_almacenados = []

# Número máximo de enlaces a extraer
max_enlaces = 300

# Bucle para realizar scroll y extraer enlaces
while len(enlaces_almacenados) < max_enlaces:
    # Hacer scroll hacia abajo
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight)')
    
    # Esperar un breve momento para que carguen los nuevos elementos (ajusta según sea necesario)
    time.sleep(2)

    # Obtener el código fuente actualizado después del scroll
    page_source = driver.page_source
    soup = BeautifulSoup(page_source, 'html.parser')

    # Encontrar nuevos enlaces y agregarlos a la lista
    enlaces = soup.select('.MuiCardContent-root.css-1qw96cp a')
    enlaces_almacenados.extend([enlace.get('href') for enlace in enlaces])

# Imprimir enlaces almacenados
for href in enlaces_almacenados[:max_enlaces]:
    print(href)

# Buscar el elemento por XPath y hacer clic
#elemento_a_clic = driver.find_element_by_xpath('//*[@id="the-app"]/div/div/div[2]/div/div/div[1]/div[3]/div/div[2]/div/div[1]')
#elemento_a_clic.click()

# Cierra el navegador
driver.quit()
```
